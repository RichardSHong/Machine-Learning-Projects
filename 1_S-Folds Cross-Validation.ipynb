3 S-Folds Cross-Validation

```python
import numpy as np 
import random

# 1. Reads in the data, ignoring the first row (header) and first column (index).
filename = "./x06Simple.csv"
data = np.loadtxt(filename, delimiter=",", skiprows=1)
#data = data[:,1:]

datanp = np.array(data).astype(int)

# (a) Randomizes the data 
np.random.seed(0)
np.random.shuffle(datanp)


X = [datanp[:,1],datanp[:,2]]
X=np.transpose(np.array(X))
y = np.array(datanp[:,3])

```

```python
# 2. 20 times does the following:
# (a) Randomizes the data 
# (b) Creates S folds.
# (c) Fori=1toS
#     i. Select fold i as your testing data and the remaining (S âˆ’ 1) folds as your training data
#     ii. Standardizes the data (except for the last column of course) based on the training data
#     iii. Train a closed-form linear regression model
#     iv. Compute the squared error for each sample in the current testing fold
# (d) You should now have N squared errors. Compute the RMSE for these.
```

```python
from sklearn.model_selection import KFold

a= datanp.shape[0]

def predict(theta, XX):
    return np.dot(XX, theta)

def getThetaClosedForm(XX,YY):
    theta = np.linalg.inv(XX.T@XX)@XX.T@YY
    return theta


folds=[3,5,20,a]
p=[]
for fold in folds:
    totalRMSEarray = []
    k=[]
    p.append(k)
    for z in range(20):
        #totalRMSE= 0
        kf = KFold(n_splits=fold, shuffle=True, random_state=z)
        q=[]
        k.append(q)
        for train_index, test_index in kf.split(X):
            #print("Fold:", fold, "TRAIN:", train_index, "TEST:", test_index)
            X_train, X_test = X[train_index], X[test_index]
            y_train, y_test = y[train_index], y[test_index]
            
            #print("TRAIN:", X_train, "TEST:", X_test)
            
    
            mean = 0
            std = 0
            std_training_data = 0
            
            mean = np.mean(X_train, axis=0)
            std = np.std(X_train, axis=0, ddof=1)
            std_training_data = (X_train-mean)/std
            
            #print("mean:",  mean, "std:", std)
            
            #Adding Bias
            std_XTrain_bias=[]
            for i in std_training_data:
                std_XTrain_bias.append([1,i[0],i[1]])
                
            std_XTrain_bias = np.array(std_XTrain_bias)
            
            #print(std_XTrain_bias)


            theta = getThetaClosedForm(std_XTrain_bias,y_train)
            #print("theta:", theta)
            
            
            
            
            
            
            
            std_testing_data = (X_test-mean)/std

            #Adding Bias
            std_XTest_bias=[]
            for i in std_testing_data:
                std_XTest_bias.append([1,i[0],i[1]])

            std_XTest_bias = np.array(std_XTest_bias)

            Y_hat = predict(theta, std_XTest_bias)


            # 7. Computes the root mean squared error (RMSE): N i=1(Yi âˆ’ Yi) . 
            #where Yi is the predicted value for observation Xi.
            from sklearn.metrics import mean_squared_error
            from math import sqrt


            RMSE = sqrt(mean_squared_error(y_test, Y_hat))
            print("Fold:", fold, "RMSE",z,":", RMSE)
            q.append([fold,RMSE])
```

```python


#1. The average and standard deviation of the root mean squared error for S = 3 over the 20 different seed values..

All_RMSE_S3 = []
for i in p[0]:
    for j in i:
        All_RMSE_S3.append(j[1])        





All_RMSE_S3 = np.array(All_RMSE_S3)


S3_mean = np.mean(All_RMSE_S3, axis=0)
S3_std = np.std(All_RMSE_S3, axis=0, ddof=1)

print("S3_mean:",S3_mean,"S3_std:", S3_std)
```
